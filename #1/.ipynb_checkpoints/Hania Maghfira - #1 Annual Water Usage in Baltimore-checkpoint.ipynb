# CREATE MODEL USING ARIMA

### 1. Split the dataset into train and test sets directly

The data will be divide into 50% for testing and 50% for training.

```python
train_size = int(len(dt)*0.5)
train = dt[0:train_size]
test = dt[train_size:]
print('Data training dimension :', train.shape)
print('Data testing dimension :',test.shape)
```

Data training contains 27 rows and data testing contains 28 rows.

### 2. ARIMA Modelling (0,1,0)
Use an ARIMA(0,1,0) on the raw data may be a good starting point.

```python
from statsmodels.tsa.arima_model import ARIMA
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
arima_order = (0,1,0)
arima = ARIMA(train, order=arima_order)
arima_fit = arima.fit()
print(arima_fit.summary())
```

ARIMA modelling with parameters (0,1,0) has error output, so it needs to be calculated with ARIMA parameter (p,d,q) with stationary test, PACF plot, and ACF plot.

##### Rolling Statistic
To find the best model of ARIMA, d parameter can be found by seeing the stationaire of data.

```python
#Determine Rolling Statistics
rolmean = pd.rolling_mean(train, window=2)
rolstd = pd.rolling_std(train, window=2)
print(rolmean, rolstd)
```

```python
orig = plt.plot(train, color='blue',label='Original')
mean = plt.plot(rolmean, color='red', label='Rolling Mean')
std = plt.plot(rolstd, color='black', label='Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.show(block=False)
```

From the graph above, we can know that :
- The data has already stationaire to mean, but not variance. So it needs to be proved by Dicky-Fuller with hypothesis :<br>
  H0 : Data is not stationaire<br>
  H1 : Data is already stationaire

##### Dicky-Fuller Test

```python
#Perform Dickey-Fuller test:
from statsmodels.tsa.stattools import adfuller

print ('Results of Dickey-Fuller Test:')
dftest = adfuller(train['Baltmore city annual water use, liters per capita per day, 1885-1968'], autolag='AIC')

dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
  dfoutput['Critical Value (%s)' %key] = value
  
print(dfoutput)
```

From the result test above, we can know that :
- p-value is 0,076, which is more than alpha 0,05
- Decision : not decline H0
- Conclusion : Data is not stasionaire

#####  ACF and PACF Plot
Find the best parameter of p and q in ARIMA model.

```python
import statsmodels.graphics.tsaplots as sts
sts.plot_acf(train, lags=20) #represents q parameter or MA
sts.plot_pacf(train, lags=20) #represents p parameter or AR
plt.show()
```

From the plot above, we can know that :
- ACF plot has lag on 1, so the value of q parameter is 1
- PACF plot has lag on 1, so the value of p parameter is 1
- The parameter of ARIMA (p,d,q) is : (1,0,1)
